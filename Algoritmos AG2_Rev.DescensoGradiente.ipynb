```python
#Función de precios

def Precios(TARIFAS):

  #Número total de nodos
  N=len(TARIFAS[0])

  #Valores iniciales tabla de precios
  PRECIOS=[ [9999]*N for i in [9999]*N]
  RUTA=[ [""]*N for i in [""]*N]

  #Recorrido iterativo de los nodos
  for i in range(N-1):
    for j in range(i+1, N):
      MIN=TARIFAS[i][j]
      RUTA[i][j]=i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN=min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j]=k
        PRECIOS[i][j]=MIN

  return PRECIOS,RUTA
```

```python
PRECIOS,RUTA=Precios(TARIFAS)
Precios(TARIFAS)
```

```python
#ASIGNACION DE TAREAS: RAMIFICACION Y PODA

#Tabla de costes para n=4
COSTES=[[11,12,18,40],[14,15,13,22],[11,17,19,23],[17,14,20,28]]
```

```python
#FUNCIONES:

def valor(S,COSTES):
  VALOR=0
  for i in range(len(S)):
    VALOR += COSTES[S[i]][i]
  return VALOR

def CI(S,COSTES):
  VALOR=0
  #Valores establecidos
  for i in range(len(S)):
    VALOR += COSTES[i][S[i]]

  #Estimacion
  for i in range(len(S),len(COSTES)):
    VALOR += min([COSTES[j][i] for j in range(len(S), len(COSTES))])
  return VALOR

def CS(S,COSTES):
  VALOR=0
  #Valores establecidos
  for i in range(len(S)):
    VALOR += COSTES[i][S[i]]

  #Estimacion
  for i in range(len(S), len(COSTES)):
    VALOR += max([COSTES[j][i] for j in range(len(S), len(COSTES))])
  return VALOR
```

```python
#Generación de Hijos (función): (0,) -> (0,1), (0,2), (0,3)
def crear_hijos(NODO, N):
  HIJOS = []
  for i in range(N ):
    if i not in NODO:
      HIJOS.append({'s':NODO +(i,)})
  return HIJOS
```

```python
#Función de ramificación y poda

def ramificacion_y_poda(COSTES):
#Construcción iterativa soluciones(árbol): se asigna un agente(ramas) por etapa

  #print(COSTES)
  DIMENSION=len(COSTES)  
  MEJOR_SOLUCION=tuple(i for i in range(len(COSTES)))
  CotaSup=valor(MEJOR_SOLUCION,COSTES)
  #print("Cota Superior:", CotaSup)

  NODOS=[]
  NODOS.append({'s':(), 'ci':CI((),COSTES)})

  iteracion = 0

  while(len(NODOS) > 0):
    iteracion +=1

    nodo_prometedor=[min(NODOS, key=lambda x:x['ci']) ][0]['s']
    #print("Nodo prometedor:", nodo_prometedor)

    #Ramificación
    #Se generan los Hijos
    HIJOS=[{'s':x['s'], 'ci':CI(x['s'], COSTES)} for x in crear_hijos(nodo_prometedor, DIMENSION)]

    #Revisión de cota superior. Si se llega a una solución final nos quedamos con ella
    NODO_FINAL=[x for x in HIJOS if len(x['s']) == DIMENSION]
    if len(NODO_FINAL )>0:
      if NODO_FINAL[0]['ci'] < CotaSup:
        CotaSup=NODO_FINAL[0]['ci']
        MEJOR_SOLUCION=NODO_FINAL
 
    #Poda
    HIJOS=[x for x in HIJOS if x['ci'] < CotaSup]

    #Añade Hijos
    NODOS.extend(HIJOS)

    #Elimina nodo ramificado
    NODOS=[x for x in NODOS if x['s'] != nodo_prometedor]

  print("La solucion final es:" ,MEJOR_SOLUCION , " en " , iteracion , " iteraciones" , " para dimension: " ,DIMENSION)
```
