```python
#Torres de Hanoi
def Torres_Hanoi(N, desde, hasta):
  #N = nº de fichas
  #desde = nº poste desde
  #hasta = nº poste hasta

  if N == 1 :
    print("Lleva la ficha desde " + str(desde) + " hasta " + str(hasta))
  else:
    Torres_Hanoi(N-1, desde, 6-desde-hasta)
    print("LLeva la ficha desde " + str(desde) + " hasta " + str(hasta))
    Torres_Hanoi(N-1, 6-desde-hasta, hasta)

  # 1, 3 -> 2
  # 1, 2 -> 3
  # 2, 3 -> 1

Torres_Hanoi(3, 1, 3)
```

```python
def cambio_monedas(cantidad, sistema):
  # cantidad a cambiar
  # sistema: valor de las monedas diferente, ordenado de mayor a menor
  solucion = {}
  for v in sistema:
    monedas = cantidad // v
    solucion[v] = monedas
    cantidad -= monedas * v
    if cantidad == 0:
      return solucion

cambio_monedas(80, [25, 10, 5, 1])
```

```python
cambio_monedas(123, [25, 10, 5, 1])
```

```python
#VERIFICA NO AMENAZAS ENTRE REINAS
def es_prometedora(SOLUCION, etapa):
  for i in range(etapa+1):
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False
    #verifica diagonales
    for j in range(i+1, etapa+1):
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#TRADUCE SOLUCION A TABLERO
def escribe_solucion(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X ", end="")
      else:
        print(" - ", end="")

#PROCESO PRINCIPAL DE N-REINAS
def reinas(N, solucion=[],etapa=0):
    if len(solucion) == 0:
      solucion = [0 for i in range(N)]

    for i in range(1, N+1):
      solucion[etapa] = i
      if es_prometedora(solucion, etapa):
        if etapa == N-1:
          print(solucion)
        else:
          reinas(N, solucion, etapa+1)
      else:
        None

        solucion[etapa] = 0

reinas(4,solucion=[],etapa=0)
```
